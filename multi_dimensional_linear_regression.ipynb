```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("hiring.csv")
df
```

```python
median1 = df["test_score(out of 10)"].median()
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(median1)
# df["Experience"]
df["Experience"] = df["Experience"].apply(w2n.word_to_num)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["Experience","test_score(out of 10)","interview score(out of 10)"]],df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# just for try
%matplotlib inline
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.plot(df["Experience"],reg.predict(df[["Experience","test_score(out of 10)","interview score(out of 10)"]]), color="red" )
plt.plot(df["test_score(out of 10)"],reg.predict(df[["Experience","test_score(out of 10)","interview score(out of 10)"]]), color="blue")
plt.plot(df["interview score(out of 10)"],reg.predict(df[["Experience","test_score(out of 10)","interview score(out of 10)"]]), color="yellow")
```
